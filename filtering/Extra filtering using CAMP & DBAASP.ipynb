```python
import os
from google.colab import drive

# --- 1. MOUNT GOOGLE DRIVE ---
drive.mount('/content/drive')

# --- 2. CONFIGURATION ---
input_file = "/content/drive/MyDrive/AMP-Generation/data/generated_amp_clustered.txt"
output_file = "/content/drive/MyDrive/AMP-Generation/data/selected_candidates.fasta"

target_numbers_str = "8, 9, 15, 17, 18, 19, 21, 22, 28, 30, 31, 34, 36, 41, 43, 44, 46, 48, 50, 51, 57, 59, 64, 66, 71, 73, 76, 81, 83, 85, 94, 95, 97, 107, 108, 109, 110, 111, 112, 113, 115, 119, 120, 121, 122, 126, 127, 129, 132, 134, 135, 136, 137, 138, 140, 142, 150, 151, 154, 156, 157, 162, 164, 168, 170, 172, 177, 179, 191, 193, 194, 198, 199, 201, 203, 207, 212, 220, 222, 224, 226, 230, 236, 238, 242, 247, 248, 250, 252, 255, 256, 257, 258, 264, 269, 272, 273, 275, 276, 281, 283, 284, 287, 289, 296, 299, 300, 304, 307, 312, 313, 315, 319, 324, 333, 337, 352, 356, 358, 360, 361, 362, 366, 367, 368, 371, 373, 383, 384, 385, 387, 394, 398, 399, 400, 405, 410, 415"

# --- 3. CONVERT STRING TO LIST OF INDICES ---
# We subtract 1 because Python lists start at 0, but your input starts at 1
target_indices = [int(x.strip()) - 1 for x in target_numbers_str.split(',')]
target_indices.sort() # Ensure we process them in order

# --- 4. FILTER AND CREATE FASTA ---
if not os.path.exists(input_file):
    print(f"Error: File not found at {input_file}")
else:
    # Use .splitlines() instead of a loop - this captures the last line
    # even if it doesn't end with a newline character.
    with open(input_file, 'r') as f:
        content = f.read()
        all_sequences = content.splitlines()

    print(f"ðŸ“Š Total sequences found in file: {len(all_sequences)}")

    selected_count = 0
    with open(output_file, 'w') as out:
        for idx in target_indices:
            if 0 <= idx < len(all_sequences):
                seq = all_sequences[idx].strip()
                if seq:  # Only write if the line isn't empty
                    out.write(f">seq_{idx + 1}\n{seq}\n")
                    selected_count += 1
            else:
                print(f"Warning: Sequence number {idx + 1} is out of range (Max is {len(all_sequences)}).")

    print(f"Success! Created {output_file}")
    print(f"Processed {selected_count} sequences into FASTA format.")
```
