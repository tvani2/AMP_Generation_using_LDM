```python
import torch
from google.colab import drive

drive.mount('/content/drive')

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

LATENT_PATH = "/content/drive/MyDrive/AMP-Generation/data/generated_latent_amp.pth"
STATS_PATH  = "/content/drive/MyDrive/AMP-Generation/models/vae_correction_stats.pth"
SAVE_PATH   = "/content/drive/MyDrive/AMP-Generation/data/generated_latent_amp_denorm.pth"

# Load generated normalized latents
z_norm = torch.load(LATENT_PATH, map_location=DEVICE)

# Load normalization stats
stats = torch.load(STATS_PATH, map_location=DEVICE)
mu_mean = stats["shift"].to(DEVICE)
mu_std  = stats["scale"].to(DEVICE)

# Denormalize
z_denorm = z_norm.to(DEVICE) * mu_std + mu_mean

# Save
torch.save(z_denorm.cpu(), SAVE_PATH)

print(f"Denormalized latents saved to: {SAVE_PATH}")
print("Shape:", z_denorm.shape)
```
